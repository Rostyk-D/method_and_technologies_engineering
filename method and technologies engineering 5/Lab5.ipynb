# Лабораторна робота #5. Методи опрацювання табличних даних за допомогою SQL запитiв.

Оберіть один зі сценаріїв, запропонованих у списку нижче. Виконайте наступні завдання:

1. Спроектуйте таблиці з необхідними полями та встановіть між ними зв'язки, якщо потрібно.
2. Заповніть таблиці тестовими даними (5-10 записів).
3. Напишіть SQL-запити для виконання різних операцій над даними.
Крім базових операцій CRUD (створення, читання, оновлення, видалення), використайте наступні ключові слова та конструкції:
* WHERE: Для фільтрації даних за певними умовами. Наприклад, знайти всіх клієнтів з певного міста.
* ORDER BY: Для сортування результатів запиту за вказаним стовпцем у зростаючому або спадаючому порядку. Наприклад, відсортувати книги за датою видання.
* GROUP BY: Для групування даних за певним стовпцем. Наприклад, підрахувати кількість книг кожного жанру.
* HAVING: Для фільтрації результатів групування. Наприклад, вивести авторів, які написали більше 5 книг.
* JOIN: Для об'єднання даних з кількох таблиць. Наприклад, пов'язати інформацію про книги з інформацією про авторів.

Після виконання всіх запитів, збережіть результати роботи у форматі CSV. Для цього ви можете використати бібліотеку Pandas.

Вся робота має бути виконана за допомогою python та SQLite3.

Варіанти сценаріїв: 
1. Бібліотека:
* Книги: id, назва, автор, жанр, рік видання
* Видавець: id, назва, країна
2. Магазин електроніки:
* Товари: id, назва, ціна, категорія
* Замовлення: id, користувач_id, товар_id, кількість
3. Ресторан:
* Страви: id, назва, ціна, категорія
* Замовлення: id, клієнт_id, страва_id, кількість
4. Фітнес-центр:
* Тренери: id, ім'я, спеціалізація
* Заняття: id, тренер_id, дата, час
5. Готель:
* Номери: id, тип, ціна
* Бронювання: id, номер_id, дата_заїзду, дата_виїзду
6. Блог:
* Статті: id, заголовок, автор, дата_публікації
* Коментарі: id, стаття_id, автор, текст
7. Онлайн-курс:
* Курси: id, назва, викладач
* Студенти: id, ім'я, курс_id
8. Музей:
* Експонати: id, назва, зал
* Відвідувачі: id, ім'я, дата_відвідування
9. Оренда автомобілів:
* Автомобілі: id, марка, модель
* Оренди: id, автомобіль_id, дата_початку, дата_кінця
10. Онлайн-ігри:
* Ігри: id, назва, жанр
* Користувачі: id, нікнейм, улюблена_гра
11. Біржа криптовалют:
* Криптовалюти: id, назва, ціна
* Транзакції: id, користувач_id, криптовалюта_id, сума
12. Соціальна мережа:
* Користувачі: id, ім'я, email
* Пости: id, користувач_id, текст
13. Система управління проектами:
* Проекти: id, назва, опис
* Завдання: id, проект_id, виконавець
14. Система керування контентом:
* Статті: id, заголовок, категорія
* Користувачі: id, ім'я, роль
15. Університет:
* Пара: id, назва, викладач, група
* Аудиторії: id, номер, пара_id
