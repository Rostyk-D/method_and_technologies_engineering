## Дані.

### Завантажте дані з файлу `Ecommerce Customers.csv` у датафрейм.

```python
df = pd.read_csv('Ecommerce Customers.csv')
df = df.drop(columns=["Email",'Address','Avatar'])
df.head()
```

### Розділіть дані на тестовий та тренувальний набори.

```python
X = df.drop(columns='Yearly Amount Spent')  # Видалити стовпець з цілею
y = df['Yearly Amount Spent']  # Цільова змінна
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Множинна лінійна регресія.

### Створіть модель та натренуйте модель.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

### Спрогнозуйте значення річних витрат для тестових даних.

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Виведення результатів
print(f"Середня квадратична помилка (MSE): {mse}")
print(f"Корінь середньоквадратичної помилки (RMSE): {rmse}")
print(f"Коефіцієнт детермінації (R^2): {r2}")
```

### Оцініть модель.

## Поліноміальна лінійна регресія.

### Спрогнозуйте значення річних витрат для тестових даних.

```python
mse = mean_squared_error(y_test, y_pred_poly)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_poly)

# Виведення результатів
print(f"Середня квадратична помилка (MSE): {mse}")
print(f"Корінь середньоквадратичної помилки (RMSE): {rmse}")
print(f"Коефіцієнт детермінації (R^2): {r2}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Список документів
documents = [
    "Машинне навчання з використанням алгоритмів",
    "Алгоритм машинного навчання для обробки даних",
    "Моделі машинного навчання для класифікації",
    "Нейронні мережі для класифікації зображень"
]

# Створюємо об'єкт TfidfVectorizer
vectorizer = TfidfVectorizer()

# Підраховуємо TF-IDF для кожного слова в кожному документі
tfidf_matrix = vectorizer.fit_transform(documents)

# Обчислюємо косинусну подібність між усіма документами
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Виводимо косинусну подібність між документами
print(cosine_sim)
```
