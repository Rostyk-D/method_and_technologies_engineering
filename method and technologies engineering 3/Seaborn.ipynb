# Seaborn

Seaborn - це бібліотека, яка дозволяє досліджувати та розуміти дані. Функції цієї бібліотеки оперують датафреймами та масивами, які містять датасети та внутрішньо виконують необхідне семантичне співставлення та статистичну агрегацію щоб отримувати інформативні графіки.

### Імпортування бібліотеки.

```python
import seaborn as sns
```

```python
import numpy as np
```

Так само, як і у matplotlib, необхідно виконати наступний рядок:

```python
%matplotlib inline
```

Дані будуть використовуватись з датасету seaborn:

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Графіки розподілів.
Є такі графіки розподілів:
* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

### [distplot](https://seaborn.pydata.org/generated/seaborn.distplot.html).
Показує графік одноварінтних наборів спостережень.

```python
sns.histplot(tips['total_bill'])
```

### [jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html). 
Дозволяє співставляти два distplot для біваріантних даних. Дозволяє обирати параметр `kind` по якому порівнювати:
* “scatter”
* “reg”
* “resid”
* “kde”
* “hex”

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

### [pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html).
Будує попарні зв’язки по всьому датафрейму (для числових стовпців) і підтримує аргумент відтінку кольору (для категоріальних стовпців)

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

### [rugplot](https://seaborn.pydata.org/generated/seaborn.rugplot.html).
Цей графік малюється таким чином, щоб для кожної точки одноварінтного розподілу була намальована позначка. Вони є будівельними блоками для kdeplot.

```python
sns.rugplot(tips['total_bill'])
```

### [kdeplot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html).
Є графіками оцінки щільності ядра. Вони замінюють кожне спостереження Гаусівським (Нормальним) розподілом з центром навколо цього значення.

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

## Категоріальні графіки.
Декілька основних типів таких графіків:
* catplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

### [barplot](https://seaborn.pydata.org/generated/seaborn.barplot.html).
__barplot__ це графік, який дозволяє агрегувати категоріальні дані по певній функції, по замовчуванню - середньому значенні:

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
sns.countplot(x = 'sex', data = tips)
```

Можна змінити об'єкт оцінювач у власну функцію, яка перетворює вектор у скаляр:

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std) # np.std -> обраховує стандартне відхилення
```

### [countplot](https://seaborn.pydata.org/generated/seaborn.countplot.html).

Графік, схожий до __barplot__, але оцінювач тут явно обраховує кількість появ. Тому тут на вхід подається тільки одне значення:

```python
sns.countplot(x = 'sex', data = tips)
```

### [boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html).
Використовуються, щоб показувати розподіл категоріальних даних. __boxplot__ показує розподіл кількісних даних шляхом порівняння змінних, або поміж рівнів категоріальних даних. Коробка показує 2 квартиль даних, вусики вказують 1 та 2 квартилі. Також є так звані точки, які визначені як `лежачі поза` використовуючи метод, який є функцією внутрішньо-квартальних меж.

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, palette = 'rainbow')
```

```python
sns.boxplot(data = tips, palette = 'rainbow', orient = 'h')
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "smoker",data = tips, palette = "coolwarm")
```

####  [violinplot](https://seaborn.pydata.org/generated/seaborn.violinplot.html).
Має схожу роль, як і boxplot. Показує розподіл кількісних даних по декількох рівнях однієї (або декількох) категоріальних змінних, таким чином, щоб ці розподіли можна було порівняти. На відміну від boxplot, у якому всі компоненти графіків відповідають точкам даних, violinplot має властивість оцінки щільності ядра підлеглого розподілу.

```python
sns.violinplot(x = "day", y = "total_bill", data = tips, palette = 'rainbow')
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips, hue = 'sex', split = True, palette = 'Set1')
```

### [stripplot](https://seaborn.pydata.org/generated/seaborn.stripplot.html) та [swarmplot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html).
__stripplot__ малює точковий графік, де серед данних є одні категоріальні. Може малюватись самостійно, але також може використовуватись разом з __boxplot__ або __violinplot__, якщо потрібно показати всі спостереження поруч з репрезентацією відповідних розподілів.

__swarmplot__ схожий до __stripplot__, але точки у ньому є скоректовані (тільки вздовж категоріальної вісі). Це дає кращу презентацію розподілів значень, хоча не добре масштабується для великого числа спостережень.

```python
sns.stripplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True)
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True, hue = 'sex', palette = 'Set1')
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True, hue = 'sex', palette = 'Set1', dodge = True)
```

```python
sns.swarmplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.swarmplot(x = "day", y = "total_bill", hue = 'sex', data = tips, palette = "Set1", dodge = True)
```

### Комбінування.
Перелічені графіки можна комбінувати між собою.

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

### [catplot](https://seaborn.pydata.org/generated/seaborn.catplot.html).

Найзагальніша форма категоріальних графіків. Може приймати параметр `kind` щоб змінювати тип графіку:

```python
sns.catplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'sex', y = 'total_bill', data = tips, kind = 'bar')
```

## Матричні графіки.
Матричні графіки дозволяють малювати дані як матриці, закодовані кольором, і вони можуть бути використані для того, щоб вказати на кластери у даних.

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.head()
```

### [heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html).
Щоб мапа тепла працювала коректно, дані мають бути у матричній формі. __sns.heatmap__ функція просто відмальовує дані у кольорі.

```python
tips.corr() # мотрична форма для даних кореляції
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), cmap = 'coolwarm', annot = True)
```

На даних про польоти:

```python
flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
```

```python
pvflights = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
sns.heatmap(pvflights)
```

```python
sns.heatmap(pvflights, cmap = 'magma', linecolor = 'white', linewidths = 1)
```

### [clustermap](https://seaborn.pydata.org/generated/seaborn.clustermap.html).
Вкористовує кластерну ієрархію щоб виробити кластеризовану версію мапи тепла.

```python
sns.clustermap(pvflights)
```

Роки та місяці більше не впорядковані, замість цього вони погруповані по схожості (у даних). Це означає, що тепер можна бачити схожості у графіку, наприклад що Липень та Серпень є схожими.

```python
 # для кращого розуміння даних можна використовувати інші методи оптимізації, наприклад нормалізацію.
sns.clustermap(pvflights, cmap = 'coolwarm', standard_scale = 1)
```
