# MatPlotLib

MatPlotLib - це бібліотека для візуалізації даних з Python.

Одні з основних перевах MatPlotLib:
* проста та зрозуміла для початку роботи;
* підтримує нестандартні етикетки та тексти;
* контроль кожного елементу у фігурі;
* високоякісні вихідні дані у різних форматах;
* підтримує багато додаткових налаштувань;

### Імпортування

```python
import matplotlib.pyplot as plt
```

### Простий приклади.

Для побудови графіків зазвичай передаються NumPy масиви або Pandas колонки. Можна передавати звичайні Python лісти.

```python
import numpy as np
x = np.linspace(-6, 6, 30)
y = x ** 2
```

Простий графік можна побудувати використовуючи наступні команди:

```python
plt.plot(x, y, 'r') # r щоб задати червоний колір графіку
plt.xlabel('Вісь Х')
plt.ylabel('Вісь Y')
plt.title('Назва графіку')
#plt.show() для інших IDE виклик цієї функції необхідний, щоб полотно вікрилось у окремому вікні
```

## MatPlotLib Об'єктно орієнтований підхід

Головна ідея у використанні об'єктно орієнтованого методу - це створити об'єкт фігури і потім викликати методи та властивості цього об'єкту. З таким підходм краще працювати, коли є декільки графіків.

```python
figure = plt.figure() # створюємо порожнє полотно
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8]) # Параметр є кортежем наступних значень (left, bottom, width, height)
axes.plot(x,y,'b') # b щоб задати синій колір графіку
axes.set_xlabel('Вісь Х')
axes.set_ylabel('Вісь Y')
axes.set_title('Назва графіку')
```

У впадку двох графіків можна зробити таке:

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.4, 0.5, 0.3, 0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel('Вісь Х 1')
axes1.set_ylabel('Вісь Y 1')
axes1.set_title('Назва графіку 1')

axes2.plot(y, x, 'r')
axes2.set_xlabel('Вісь Х 2')
axes2.set_ylabel('Вісь Y 2')
axes2.set_title('Назва графіку 2');
```

### Підграфіки.

plt.subplots() об'єкт діє як автоматичний менеджер вісей.

```python
figure, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Назва');
```

Також можна задати конкретне число рядків та стовпців коли створюється subplots() об'єкт.

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

Через масив вісей можна проітерувати та побудувати потрібні графіки:

```python
for ax in axes:
    ax.plot(x,y,'g')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Назва')

figure # виклик полотна, щоб показати графіки
```

### Легенди

Для того, щоб додати легенду на графік, потрібно використати аргумент __label="Текст"__ коли додаєтьсяг рафік, і потім використати метод legend().

Метод legend() приймає параметр loc, що визначає де на графіку розташувати легенду. Значення цього параметра наступні:
* 0 => matplotlib сам вирішує, яка локація є оптимальною для легенди;
* 1 => верхній правий кут;
* 2 => верхній лівий кут;
* 3 => нижній лівий кут;
* 4 => нижній правий кут.

```python
figure = plt.figure()

axes = figure.add_axes([0,0,1,1])

axes.plot(x, x**2, label="x**2")
axes.plot(x, x**3, label="x**3")
axes.legend(loc=2)
```

### Кольори, ширина та типи ліній.

matplotlib дозволяє налаштовувати різні параметри графіків.

#### Колір
Існують такі способи налаштування кольору:
1. у методі ax.plot(x,x**,'b') буква у останньому параметрі відповідає за задання кольору;
2. з допомогою параметру color=: ax.plot(x, x+2, color="#8B008B"). Як видно, не обов'язково задавати колір буквою, а можна напряму задавати колір у форматі RGB.

#### Лінії та типи маркерів

Щоб змінити ширину лінії можна використати параметр `linewidth` або `lw`. Стиль можна обрати з допомогою параметру `linestyle` або `ls`: 

```python
figure, axes = plt.subplots(figsize=(12,6))

axes.plot(x, x+1, color="red", linewidth=0.25)
axes.plot(x, x+2, color="red", linewidth=0.50)
axes.plot(x, x+3, color="red", linewidth=1.00)
axes.plot(x, x+4, color="red", linewidth=2.00)

axes.plot(x, x+5, color="green", lw=3, linestyle='-')
axes.plot(x, x+6, color="green", lw=3, ls='-.')
axes.plot(x, x+7, color="green", lw=3, ls=':')

line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # формат: line length, space length, ...

axes.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
axes.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
axes.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
axes.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

axes.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
axes.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
axes.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Межі графіку.

Можна налаштовувати межі вісей використовуючи методи `set_ylim` i `set_xlim` об'єкту вісей, або `axis('tight')` для автоматичного налаштування вісей:

```python
figure, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Межі вісей по замовчуванню")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Тісні осі")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Налаштовані межі осей");
```
